#### Chart1/ Question 1- What is relationship between the price of AirBnb listing and number of reviews it has?

````python
#First we remove outliers so we can get a more understandable visualization
df_filtered=df.copy()

# Calculate the IQR for price variable
Q1=df_filtered["price"].quantile(0.25)
Q3=df_filtered["price"].quantile(0.75)
IQR=Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Remove any data points that fall outside the IQR range
df_filtered = df_filtered[(df_filtered['price'] >= lower_bound) & (df_filtered['price'] <= upper_bound)]
print(df_filtered.columns)
````

````python
#Creating scatter plot showing relationship betn price and no. of reviews
plt.scatter(df_filtered["price"],df_filtered["number_of_reviews"])
plt.xlabel("Price")
plt.ylabel("Number of Reviews")
plt.title("Price Vs Number of Reviews(Outliers removed)")
plt.show()
````

##### 1. Why did you pick the specific chart?


Scatter plot is used here because it visualizes and analyze the relationship between two variables, help identifying potential correlation, trends and pattern in data.

##### 2. What is/are the insight(s) found from the chart?

**Insights**:

* Based on results of scatter plot we can conclude that there is positive correlation between price and number of reviews it has even after removing outliers.

* Listings with more reviews tend to be more expensive.

However,there are other factors that influence the price of AirBnb listings such as location of listing , the amenities that it offers and time of year.

#### Chart - 3/Question 3- How does the number of Airbnb bookings vary by neighborhood and time of year?

````python
# Chart - 3 visualization code
#Converting the "last_review" column to datetime object
df["last_review"]=pd.to_datetime(df["last_review"])

# Extracting the month from the "last_review" column
df['month'] = df['last_review'].dt.month

# Calculating the number of Airbnb bookings for each month and neighborhood
num_bookings_by_month_and_neighborhood = df.groupby(['month', 'neighbourhood'])['id'].count()

# Unstacking the DataFrame
num_bookings_by_month_and_neighborhood = num_bookings_by_month_and_neighborhood.unstack()

# Converting the DataFrame to a Series
num_bookings_by_month_and_neighborhood = num_bookings_by_month_and_neighborhood.sum(axis=1)
````

````python
# Visualizing the results by plotting bar chart
plt.bar(num_bookings_by_month_and_neighborhood.index, num_bookings_by_month_and_neighborhood.values)
plt.xlabel('Month')
plt.ylabel('Number of bookings')
plt.title('Number of Airbnb Bookings by Month (NYC Airbnb Bookings, 2019)')
plt.show()
````

##### 1. Why did you pick the specific chart?

Here we choose bar graph to visually compare between the categorical value- 'Month' and its numerical value- 'Bookings count'.The length of bars is directly corresponds to values, making it easy to visually compare and identify significant differences between categories.

##### 2. What is/are the insight(s) found from the chart?

Insights :

* The number of Airbnb bookings varies significantly by month. In general, there are more Airbnb bookings in the summer months (April , May , June, July) than in the winter months (December, January, February , March).

This is likely due to a number of factors, including:

1)The weather is more pleasant in the summer months, making it a more attractive time to visit New York City.

2)People are more likely to take vacations during the summer months.

Are there any insights that lead to negative growth? Justify with specific reason.
